```python
import random
###creating player
def create_player(name):
    player = {}
    player['Name'] = name
    player['Level'] = 1
    player['XP'] = 0
    player['HP'] = 100
    player['Attack'] = 10
    player['Defence'] = 5
    return print(f"Hello {name} welcome to the game \n your level is {player['Level']} \n Stats: \n HP: {player['HP']} \n Attack: {player['Attack']} \n Defence: {player['Defence']}")
    

###create inventory
def create_inventory():
    return {}

#Add to inventory
def add_to_inventory(inventory,item):
    print("Opening inventory")
    print("==== " * 10)
    inventory[item] = 1
    print(f"{item} has been added to your inventory")
    return inventory

#view inventory
def view_inventory(inventory):
    for index, key in enumerate(inventory.items()):
        print(f"{index + 1}: {key} ")
 

def weapon_selection(choice):
    weapon_choices = list(weapons.keys())
    weapon_name = weapon_choices[choice]
    weapon = weapons[weapon_name]
    

    print(f"You hgave gained the {weapon_name}")
    print(f"Stats:")

    for stat, value in weapon.items():
        print(f"{stat} : {value}")
    return weapon_name
  

def encounter_monster():
    monster_name = random.choice(list(monsters.keys()))
    monster = monsters[monster_name]
    print(f"You have encountered a level {monster['level']} {monster_name} with {monster['hp']} hp")


name = input("Enter your name")
create_player(name)
bag = create_inventory()
choice = int(input(f"Please pick a number between 0 and {len(weapons)}"))
weapon = weapon_selection(choice)
add_to_inventory(bag,weapon)

choice = input("Would you like to view your inventory? y/n")
if choice == "y":
    view_inventory(bag)
else:
    pass


encounter_monster()

choice = int(input(f"What do you want to do? \n 1. Attack \n 2. Run \n 3. Inventory \n"))
```
